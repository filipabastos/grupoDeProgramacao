**EXERCICIO 4**

Calcular para cada produto o total das vendas do ano e mostre-os usando Pie Chart (gráfico de queijo).
No Pie chart inscrever a % de vendas de cada produto.

```python
sumVendCreme = df["facecream"].sum()
sumVendSabaoFacial = df["facewash"].sum()
sumVendPasta = df["toothpaste"].sum()
sumVendSabonete = df["bathingsoap"].sum()
sumVendShampoo = df["shampoo"].sum()
sumVendHidratante = df["moisturizer"].sum()

totalunits = df["total_units"].sum()
totVendCreme = (sumVendCreme / totalunits) 
totVendSabaoFacial = (sumVendSabaoFacial / totalunits) 
totVendPasta = (sumVendPasta / totalunits) 
totVendSabonete = (sumVendSabonete / totalunits) 
totVendShampoo = (sumVendShampoo / totalunits) 
totVendHidratante = (sumVendHidratante / totalunits)


dados = [totVendCreme,totVendSabaoFacial,totVendPasta,totVendSabonete,totVendShampoo,totVendHidratante]
rotulos = ["Creme Facial","Sabão Facial","Pasta de Dentes","Sabonete","Shampoo","Creme Hidratante"]

plt.axis("equal")
plt.pie(dados,labels = rotulos,radius=1.5,autopct = "%0.2f%%",shadow=False, startangle=0)

print("Resposta")
plt.show()
```
