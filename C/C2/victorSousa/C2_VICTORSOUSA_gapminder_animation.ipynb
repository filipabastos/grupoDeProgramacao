### Análise dos Códigos:

**Tem como função importar as bibliotecas, analisar e visualizar os dados**

```python
# libraries
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
```

**Importa o ficheiro csv do url e em sequência ler com a biblioteca pandas e confirma a sua correta importação através do head que por defaul mostra as 5 primeiras linhas**

```python
# Get the data (csv file is hosted on the web)
url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv'
data = pd.read_csv(url)
#data=pd.read_csv('gapminderData.csv')
data.head()
```

**Tem como função através do "pd.Categorical(data['continent'])" ler as informações da coluna e "data['continent']=pd.Categorical(data['continent'])" faz a contagem dos continentes separando as as informações em 5 valores**

```python
# And I need to transform my categorical column (continent) in a numerical value group1->1, group2->2...
data['continent']=pd.Categorical(data['continent'])
```

**A contrução foi feita através do método plot scatter. Feiot para ver os anos nos cinco continentes, a evolução da relação do GDP per Capita e o Life Expectancy que é apresentado pelo gráfico por meio do ciclo "for".**

**Ao iniciar o matplotlib é definido um clico "for" para percorreer os vários anos da coluna "year", a largura dos eixos(linha de código: flg:=plt.figure(flgsize=(680/my_dpi,480/my_dpl), dpl=my_dpl)).**

**Foi definido posteriormente um tempo para apresentar os anos e o decorrer, tendo como base(tmp=[data.year ==i]). Apresentado como plt.scatter.**

**Definido os eixos "y" com o "com'gdpPercap', e o eixo "x" com 'lifeExp' o parametro "s" define o tamanho do marcardor que será definido para "coluna['pop']/2000, c= é responsável pela coloração que representa os 5 continentes, os restantes parametros relacionam'se com as características físicas da bolha.**

**Ciclo For: Tem como função ser uma variável dos anos para se obter as informações sobre "GDP per Capita" e "life Expentacy" de acordo com cada continente apresentado.**




```python
%matplotlib notebook
%matplotlib notebook
 
# For each year:
for i in data.year.unique():
 
# initialize a figure
    fig=plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
 
# Change color with c and alpha. I map the color to the X axis value.
    tmp=data[ data.year == i ]
    plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
 
# Add titles (main and on axis)
    plt.yscale('log')
    plt.xlabel("Life Expectancy")
    plt.ylabel("GDP per Capita")
    plt.title("Year: "+str(i) )
    plt.ylim(0,100000)
    plt.xlim(30, 90)
 
# Save it
    filename='Gapminder_step'+str(i)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()

#Transformation in a GIF using Image Magick

# Then use image magick (this is bash, not python)
#magick convert -delay 80 Gapminder*.png animated_gapminder.gif
```
